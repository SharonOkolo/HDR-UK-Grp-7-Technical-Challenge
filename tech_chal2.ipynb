```python
#Merge regimen → outcome
df = pd.merge(
    outcome, 
    regimen[['MERGED_REGIMEN_ID', 'ENCORE_PATIENT_ID', 'DATE_DECISION_TO_TREAT', 'START_DATE_OF_REGIMEN']], 
    on='MERGED_REGIMEN_ID', 
    how='left'
)

```

```python
# patient demographics
df = pd.merge(
    df, 
    patient[['PATIENTID', 'ETHNICITY', 'GENDER']], 
    left_on='ENCORE_PATIENT_ID', 
    right_on='PATIENTID', 
    how='left'
)
```

```python
#Merge deprivation info from tumour
df = pd.merge(
    df, 
    tumour[['PATIENTID', 'QUINTILE_2019']], 
    on='PATIENTID', 
    how='left'
)
```

```python
#contingency test for each variable
# Ethnicity vs outcome
ethnicity_outcome_ct = pd.crosstab(df['ETHNICITY'], df['REGIMEN_OUTCOME_SUMMARY'])

# Gender vs outcome
gender_outcome_ct = pd.crosstab(df['GENDER'], df['REGIMEN_OUTCOME_SUMMARY'])

# Deprivation vs outcome
deprivation_outcome_ct = pd.crosstab(df['QUINTILE_2019'], df['REGIMEN_OUTCOME_SUMMARY'])
```

```python
#chi test of independence 
from scipy.stats import chi2_contingency

# Ethnicity vs outcome
chi2_eth, p_eth, dof_eth, expected_eth = chi2_contingency(ethnicity_outcome_ct)

# Gender vs outcome
chi2_gender, p_gender, dof_gender, expected_gender = chi2_contingency(gender_outcome_ct)

# Deprivation vs outcome
chi2_dep, p_dep, dof_dep, expected_dep = chi2_contingency(deprivation_outcome_ct)
```

```python
print(f"Ethnicity vs Outcome p-value: {p_eth:.4f}")
print(f"Gender vs Outcome p-value: {p_gender:.4f}")
print(f"Deprivation vs Outcome p-value: {p_dep:.4f}")
```

```python
# Recreate model data from scratch
df_model = df[['OUTCOME_BINARY', 'ETHNICITY', 'GENDER', 'QUINTILE_2019']].copy()

# Drop any rows with missing values
df_model.dropna(inplace=True)

# Double-check column types before encoding
print(df_model.dtypes)
```

```python
# One-hot encode categorical variables
df_model_encoded = pd.get_dummies(df_model, columns=['ETHNICITY', 'GENDER', 'QUINTILE_2019'], drop_first=True)

# Ensure all columns are numeric
print(df_model_encoded.dtypes)
```

```python
import statsmodels.api as sm
import numpy as np

X = df_model_encoded.drop(columns=['OUTCOME_BINARY'])
y = df_model_encoded['OUTCOME_BINARY']

# Add constant and ensure numeric arrays
X = sm.add_constant(X)
X = X.astype(float)
y = y.astype(float)

# Run logistic regression
model = sm.Logit(y, X).fit()
print(model.summary())
```

```python
odds_ratios = np.exp(model.params)
conf_int = np.exp(model.conf_int())
results = pd.DataFrame({
    'Odds Ratio': odds_ratios,
    'CI Lower': conf_int[0],
    'CI Upper': conf_int[1],
    'P-value': model.pvalues
})
print(results)
```
